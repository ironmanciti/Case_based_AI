# 3. Matplotlib Crash - Basic Matplotlib

### 선긋기

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [5,8,10]
y = [12,16,6]
x2 = [6, 9, 11]
y2 = [6, 15, 7]

plt.plot(x,y)
plt.plot(x2,y2)

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

### bar chart

```python
plt.bar(x, y)
plt.bar(x2, y2)

plt.title('bar chart')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

### scatter plot (산점도)

```python
plt.scatter(x, y, label='x')
plt.scatter(x2, y2, label='x2')

plt.title('Scatter plot')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
```

## histogram

```python
x = np.random.randn(1000)

plt.hist(x, bins=10)

plt.title('histogram')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

## Imshow

image data 처럼 행과 열을 가진 행렬 형태의 2차원 데이터는 imshow 로 표시

```python
from sklearn.datasets import load_digits
digits = load_digits()
X = digits.images[0]
X
```

```python
plt.title("mnist digits; 0")
plt.imshow(X, interpolation='nearest', cmap='gray')
plt.xticks([])
plt.yticks([])
plt.grid(False)
```
